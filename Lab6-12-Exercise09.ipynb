# Lab 3 Exercise 9



## Using StringBuilder
จากการทดลองที่ผ่านมา เราจะเห็นว่า string เป็น object ชนิด immutable คือไม่สามารถแก้ไขได้ การแก้ไขใดๆ ใน string จะส่งผลให้ระบบสร้าง string ใหม่ขึ้นมาแทน string เก่า

ในการใช้งานที่มีปริมาณการกระทำต่อ string ซ้ำ ๆ เป็นจำนวนมาก อาจจะก่อให้เกิดปัญหาด้านประสิทธิภาพการทำงานของระบบคอมพิวเตอร์ได้

คลาส StringBuilder จะมีการสร้าง buffer ของ string ไว้ภายใน ทำให้ไม่ต้องสร้าง string ใหม่เมื่อมีการเปลี่ยนแปลงค่าในข้อความดังเช่นการใช้ string โดยทั่วไป

>       sb.GetHashCode = {sb.GetHashCode()}");
>   ```

3. อธิบายโปรแกรม

ให้ศึกษาโปรแกรมด้านบนและอธิบายการทำงาน

ตอบ โปรแกรมนี้ใช้คลาส StringBuilder จาก System.Text ใน C# เพื่อทำการสร้างและแก้ไขสตริงที่มีประสิทธิภาพมากขึ้น โดยการใช้ StringBuilder นั้นสามารถเปลี่ยนแปลงเนื้อหาของสตริงได้โดยไม่ต้องสร้างออบเจ็กต์ใหม่ทุกครั้ง ซึ่งแตกต่างจากการใช้ string ปกติใน C# ที่จะไม่สามารถแก้ไขเนื้อหาของสตริงได้โดยตรง (เนื่องจาก string เป็น immutable). 
โปรแกรมนี้สร้าง StringBuilder และใช้เมธอด GetHashCode() เพื่อดูรหัสแฮชของออบเจ็กต์ก่อนและหลังการเปลี่ยนแปลง.
StringBuilder สามารถทำการเปลี่ยนแปลงเนื้อหาของข้อความได้ (ในกรณีนี้, เปลี่ยน "R" เป็น "C"), ซึ่งแตกต่างจาก string ที่ไม่สามารถเปลี่ยนแปลงได้ (immutable).
การเปลี่ยนแปลงนี้จะทำให้ GetHashCode() ของ StringBuilder เปลี่ยนแปลงเนื่องจากเนื้อหาของมันได้ถูกแก้ไข.

6. อธิบายโปรแกรม

ให้ศึกษาโปรแกรมด้านบนและอธิบายการทำงาน

โปรแกรมนี้ใช้ StringBuilder เพื่อสร้างข้อความที่ประกอบด้วยตัวเลข 0 ถึง 9 ในลูปและแสดงข้อมูลต่าง ๆ เกี่ยวกับ StringBuilder ในแต่ละรอบของลูป รวมถึงแสดงผลลัพธ์สุดท้ายหลังจากที่เสร็จสิ้นการทำงานทั้งหมด.
โปรแกรมนี้ใช้ StringBuilder เพื่อสร้างข้อความที่มีตัวเลขตั้งแต่ 0 ถึง 9 โดยการเพิ่มตัวเลขลงใน StringBuilder ในแต่ละรอบของลูป.
มันแสดงข้อมูลของ StringBuilder หลังจากการเพิ่มตัวเลขในแต่ละรอบ รวมถึงการแสดง GetHashCode() ซึ่งจะเปลี่ยนแปลงทุกครั้งที่มีการเปลี่ยนแปลงเนื้อหาของ StringBuilder.
ผลลัพธ์สุดท้ายที่ได้จะเป็นข้อความที่มีตัวเลขจาก 0 ถึง 9 รวมกัน.
